```python
%matplotlib inline
from matplotlib import pyplot as plt
import csv
f = open("GenderDiscrimination.csv")
salary_list = [row for row in csv.reader(f)][1:]
f.close()
counts = dict()
gender = [x[0] for x in salary_list]
years = [int(x[1]) for x in salary_list]
salary =[int(x[2]) for x in salary_list]
female_years = years[:129]
male_years = years[130:]
female_salary = salary[:129]
male_salary = salary[130:]
for item in salary_list:
    counts[item[0]] = counts.get(item[0], 0) + 1
print counts

female = plt.scatter(female_years, female_salary, marker='x', color='#f660ab')
male = plt.scatter(male_years, male_salary, color='#000080')
plt.title("Gender Discrimination")
plt.ylabel("Salary per annum(USD)")
plt.xlabel("Years of Experience")
plt.legend((female, male), ("Female", "Male"))
plt.show
```
